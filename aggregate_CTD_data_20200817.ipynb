```python
infile = '../Data/CTD_data/Castaway_ODU/CastAway_ODU_concat.txt'
castaway = pd.read_csv(infile)
castaway.head()
```

```python
infile = '../Data/CTD_data/RBR_data_geolocation_annotation.csv'
rbr_geo = pd.read_csv(infile)

# s = pd.Series(pd.to_datetime(rbr_geo['Time 1']))
# rbr_geo['DateTime_rounded'] = s.dt.round('min')

rbr_geo.head(20)
```

```python
infile = '../Data/CTD_data/RBR_data.csv'
rbr = pd.read_csv(infile)

# we only need RBR data from 8/5/19 and 8/16/19
rbr = rbr[(rbr.Year == 2019) & (rbr.Month == 8) & (rbr.Day == 5) | (rbr.Day == 16)]

rbr.head()
```

```python
lat = []
lon = []
hr = []
for i in rbr.Counter:
    year = rbr.Year[rbr.Counter == i].item()
    month = rbr.Month[rbr.Counter == i].item()
    day = rbr.Day[rbr.Counter == i].item()
    hour = rbr.Hour_EST[rbr.Counter == i].item()
    min_h = rbr.Min_EST[rbr.Counter == i].item() + 3
    min_l = rbr.Min_EST[rbr.Counter == i].item() - 3
    
    lat_i = rbr_geo.Latitude[(rbr_geo.Year == year) & (rbr_geo.Month == month) & (rbr_geo.Day == day) & (rbr_geo.Hour_EST == hour) & (rbr_geo.Min_EST < min_h) & (rbr_geo.Min_EST > min_l)]
    lon_i = rbr_geo.Longitude[(rbr_geo.Year == year) & (rbr_geo.Month == month) & (rbr_geo.Day == day) & (rbr_geo.Hour_EST == hour) & (rbr_geo.Min_EST < min_h) & (rbr_geo.Min_EST > min_l)]

    hr.append(hour - 4) # rbr data is in GMT 
    
    if len(lat_i) < 1:
        lat_i = "NaN"
        lon_i = "NaN"
    else:
        lat_i = lat_i.item()
        lon_i = lon_i.item()
        
    lat.append(lat_i)
    lon.append(lon_i)
```

```python
listit = np.arange(1,len(rbr.Year)+1)
Instrument = []
for i in listit:
    Instrument.append('RBR')
    
rbr_new = pd.DataFrame({'Instrument':Instrument,'Year':rbr.Year,'Month':rbr.Month,'Day':rbr.Day,'Hour_EST':hr,'Min_EST':rbr.Min_EST,'Lat':lat,'Lon':lon,'Depth':rbr.Depth,'Temp':rbr.Temperature,'Sal':rbr.Salinity})
```

```python
listit = np.arange(1,len(castaway.Year)+1)
Instrument = []
for i in listit:
    Instrument.append('CastAway')
    
castaway_new = pd.DataFrame({'Instrument':Instrument,'Year':castaway.Year,'Month':castaway.Month,'Day':castaway.Day,'Hour_EST':castaway.Hour_EST,'Min_EST':castaway.Min_EST,'Lat':castaway.Lat,'Lon':castaway.Lon,'Depth':castaway.Depth_m,'Temp':castaway.Temperature_C,'Sal':castaway.Salinity})
```

```python
frames = [rbr_new,castaway_new]
df = pd.concat(frames, sort = False)

df.to_csv('../Calculations/CTD_Data_All_concat.csv')

df.head()
```

```python
infile = "../Data/OBX_Spl_Date_Times.csv"
datetime = pd.read_csv(infile)
datetime.head()
```

```python
infile = "../Calculations/OBX_Chla_compiled_20200817.csv"
chl = pd.read_csv(infile)
chl.head()
```

```python
listit = np.arange(1,19)

lat = []
lon = []
sst = []
sss = []
data_source = []
loc_source = []
truehour = []
truemin = []
chl_mean = []
chl_sd = []

for i in listit:
    year = datetime.Year[datetime.Counter == i].item()
    month = datetime.Month[datetime.Counter == i].item()
    day = datetime.Day[datetime.Counter == i].item()
    event = datetime.Event[datetime.Counter == i].item()
    loc = datetime.Location[datetime.Counter == i].item()
    
    if event == "POINT":
        ctd = df[(df.Year == year) & (df.Month == month) & (df.Day == day)]
        chl_i = np.mean(chl.Chla_ugL[(chl.Year == year) & (chl.Month == month) & (chl.Day == day)])
        chl_sd_i = np.std(chl.Chla_ugL[(chl.Year == year) & (chl.Month == month) & (chl.Day == day)])
        
    else:
        hour = datetime.Approx_Hour_EST[datetime.Counter == i].item()
        min_h = datetime.Approx_Min_EST[datetime.Counter == i].item() + 20
        min_l = datetime.Approx_Min_EST[datetime.Counter == i].item() - 20
        ctd = df[(df.Year == year) & (df.Month == month) & (df.Day == day) & (df.Hour_EST == hour) & (df.Min_EST < min_h) & (df.Min_EST > min_l)]
        chl_i = np.mean(chl.Chla_ugL[(chl.Year == year) & (chl.Month == month) & (chl.Day == day) & (chl.Station == loc)])
        chl_sd_i = np.std(chl.Chla_ugL[(chl.Year == year) & (chl.Month == month) & (chl.Day == day) & (chl.Station == loc)])
  
        if len(ctd) == 0:
            hour = datetime.Approx_Hour_EST[datetime.Counter == i].item() - 1
            min_h = 60
            min_l = 40
            ctd = df[(df.Year == year) & (df.Month == month) & (df.Day == day) & (df.Hour_EST == hour) & (df.Min_EST < min_h) & (df.Min_EST > min_l)]
        
    lat_i = np.mean(ctd.Lat[ctd.Lat != 'NaN'])
    lon_i = np.mean(ctd.Lon[ctd.Lon != 'NaN'])
    sst_i = np.mean(ctd.Temp[(ctd.Sal > 20) & (ctd.Depth < 1)]) # note: values may be odd when instrument is turned on (outside of water); remove any values with anomalously low salinities 
    sss_i= np.mean(ctd.Sal[(ctd.Sal > 20) & (ctd.Depth < 1)])
    truehour_i = np.mean(ctd.Hour_EST)  
    truemin_i = np.mean(ctd.Min_EST)  
  
    data_source_i = ctd.Instrument.unique()
    if len(data_source_i) > 0:
        data_source_i = data_source_i.item()
    else:
        data_source_i = 'NaN'
    
    if np.isnan(lat_i) == False:
        loc_source_i = 'CTD'
    else:
        loc_source_i = 'NaN'
        
    lat.append(lat_i)
    lon.append(lon_i)
    sst.append(sst_i)
    sss.append(sss_i)
    data_source.append(data_source_i)
    loc_source.append(loc_source_i)
    truehour.append(truehour_i)
    truemin.append(truemin_i)
    chl_mean.append(chl_i)
    chl_sd.append(chl_sd_i)
```

```python
infile = "../Data/OBX_Spl_Date_Times.csv"
datetime = pd.read_csv(infile)
datetime.head()

datetime['Loc_Data_Source'] = loc_source
datetime['Lat'] = lat
datetime['Lon'] = lon
datetime['Hydro_Data_Source'] = data_source
datetime['CTD_Hour_EST'] = truehour
datetime['CTD_Min_EST'] = truemin
datetime['Sfc_Temp_C'] = sst
datetime['Sfc_Sal'] = sss
datetime['Sfc_Chla_ugL'] = chl_mean
datetime['Sfc_Chla_ugL_stdev'] = chl_sd

datetime.to_csv('../Calculations/OBX_Compiled_Sfc_Data_1.csv')

print(datetime)
```
